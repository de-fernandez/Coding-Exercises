Given three integers a, b, and c, return the largest number obtained after inserting the operators +, *, and parentheses (). In other words, try every combination of a, b, and c with the operators, without reordering the operands, and return the maximum value.

Example<br>
With the numbers 1, 2, and 3, here are some possible expressions:

1 * (2 + 3) = 5<br>
1 * 2 * 3 = 6<br>
1 + 2 * 3 = 7<br>
(1 + 2) * 3 = 9<br>
The maximum value that can be obtained is 9.

Notes<br>
The numbers are always positive, in the range 1 ≤ a, b, c ≤ 10.<br>
You can use the same operation more than once.<br>
It is not necessary to use all the operators or parentheses.<br>
You cannot swap the operands. For example, with the given numbers, you cannot get the expression (1 + 3) * 2 = 8.

Input and Output Examples<br>
expressionsMatter(1, 2, 3) ==> 9, because (1 + 2) * 3 = 9.<br>
expressionsMatter(1, 1, 1) ==> 3, because 1 + 1 + 1 = 3.<br>
expressionsMatter(9, 1, 1) ==> 18, because 9 * (1 + 1) = 18.<br>

```python
def expression_matter(a, b, c):
    exp1 = a + b + c
    exp2 = a * b * c
    exp3 = (a + b) * c
    exp4 = a * (b + c)
    exp5 = a + b * c
    exp6 = a * b + c

    largest = exp1

    if exp2 > largest:
        largest = exp2
    if exp3 > largest:
        largest = exp3
    if exp4 > largest:
        largest = exp4
    if exp5 > largest:
        largest = exp5
    if exp6 > largest:
        largest = exp6

    return largest
```

```python
expression_matter(1, 8, 3) # 27
```
