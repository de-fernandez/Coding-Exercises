Kids drink toddy.<br>
Teens drink coke.<br>
Young adults drink beer.<br>
Adults drink whisky.<br>
Make a function that receive age, and return what they drink.

Rules:<br>
Children under 14 old.<br>
Teens under 18 old.<br>
Young under 21 old.<br>
Adults have 21 or more.<br>
Examples: (Input --> Output)

13 --> "drink toddy"<br>
17 --> "drink coke"<br>
18 --> "drink beer"<br>
20 --> "drink beer"<br>
30 --> "drink whisky"


```python
def people_with_age_drink(age):
    if age < 14:
        return "drink toddy"
    elif age < 18:
        return "drink coke"
    elif age < 21:
        return "drink beer"
    else:
        return "drink whisky"
```

```python
people_with_age_drink(14) # 'drink coke'
```
