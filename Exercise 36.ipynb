Write a function which calculates the average of the numbers in a given array.

Note: Empty arrays should return 0.

```python
def find_average(numbers):
    
    if len(numbers) == 0:
        return 0
    
    result = 0
    for i in numbers:
        result += i
    return (result/len(numbers))
```

```python
find_average([1, 2, 3]) # 2 
```
