A child drops a ball from height h.

The ball bounces to a fraction of its previous height (bounce).

A mother watches from a window at height window.

Count how many times the ball passes the window.

Rules:<br>
- h > 0<br>
- 0 < bounce < 1<br>
- window < h<br>
If not valid → return -1

The ball is seen:<br>
- once when falling past the window<br>
- once when rising past the window (if it bounces high enough)

The ball keeps bouncing until it no longer goes above the window.

Return total number of times the ball is seen.

```python
def bouncing_ball(h, bounce, window):
    if h <= 0 or bounce <= 0 or bounce >= 1 or window >= h:
        return -1

    count = 0

    while h > window:
        count += 1
        h *= bounce

        if h > window:
            count += 1

    return count
```

```python
bouncing_ball(3, 0.66, 1.5) # 3
```
