The number 89 is the first integer with more than one digit that fulfills the property partially introduced in the title of this kata.

What's the use of saying "Eureka"? Because this sum gives the same number:

89 = 8¹ + 9²

The next number having this property is 135:

135 = 1¹ + 3² + 5³

Task

We need a function to collect these numbers. The function receives two integers, a and b, which define the range [a, b] (inclusive), and outputs a list of the sorted numbers in the range that fulfill the property described above.

Examples

1, 10  --> [1, 2, 3, 4, 5, 6, 7, 8, 9]<br>
1, 100 --> [1, 2, 3, 4, 5, 6, 7, 8, 9, 89]

If there are no numbers of this kind in the range [a, b], the function should output an empty list.

90, 100 --> []

Enjoy it!!

```python
def sum_dig_pow(a, b):
    result = []
    sum = 0
    count = 1
    for num in range(a, b + 1):
        for char in str(num):
            sum += (int(char) ** count)
            count += 1
        if sum == num:
            result.append(num)
        sum = 0
        count = 1
    return result
```

```python
sum_dig_pow(1, 100) # [1, 2, 3, 4, 5, 6, 7, 8, 9, 89]
```
