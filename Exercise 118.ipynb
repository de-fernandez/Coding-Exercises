Your job is to write a function which increments a string, to create a new string.

If the string already ends with a number, the number should be incremented by 1.<br>
If the string does not end with a number. the number 1 should be appended to the new string.

Examples:

foo -> foo1

foobar23 -> foobar24

foo0042 -> foo0043

foo9 -> foo10

foo099 -> foo100

Attention: If the number has leading zeros the amount of digits should be considered.

```python
def increment_string(string):
    container = ""
    for char in reversed(string):
        if char.isdigit():
            container = char + container
        else:
            break

    if container == "":
        return string + "1"

    result = string[:-len(container)]
    length = len(container)
    container = str(int(container) + 1).zfill(length)

    return result + container
```

```python
increment_string("fo99obar99") # fo99obar100
```
