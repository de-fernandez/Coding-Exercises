Given an array of ones and zeroes, convert the equivalent binary value to an integer.

Eg: [0, 0, 0, 1] is treated as 0001 which is the binary representation of 1.

Examples:

Testing: [0, 0, 0, 1] ==> 1<br>
Testing: [0, 0, 1, 0] ==> 2<br>
Testing: [0, 1, 0, 1] ==> 5<br>
Testing: [1, 0, 0, 1] ==> 9<br>
Testing: [0, 0, 1, 0] ==> 2<br>
Testing: [0, 1, 1, 0] ==> 6<br>
Testing: [1, 1, 1, 1] ==> 15<br>
Testing: [1, 0, 1, 1] ==> 11<br>
However, the arrays can have varying lengths, not just limited to 4.


```python
def binary_array_to_number(arr):
    binary = {0: 1}
    result = 0
    power = len(arr) - 1
    for i in range(len(arr)):
        binary[i] = 2 ** power
        power -= 1
    for key, value in binary.items():
        result = result + arr[key]* value
    return result
        
```

```python
binary_array_to_number([1, 0, 1, 1]) # 11
```
