Given two arrays a and b, write a function comp(a, b) (or compSame(a, b)) that checks whether the two arrays have the "same" elements, with the same multiplicities (the multiplicity of a member is the number of times it appears).

"Same" means, here, that the elements in b are the elements in a squared, regardless of the order.

Examples

Valid arrays

a = [121, 144, 19, 161, 19, 144, 19, 11]<br>
b = [121, 14641, 20736, 361, 25921, 361, 20736, 361]

comp(a, b) returns true because in b:<br>
121 is the square of 11,<br>
14641 is the square of 121,<br>
20736 is the square of 144,<br>
361 is the square of 19,<br>
25921 is the square of 161,<br>
and so on.

It gets obvious if we write b's elements in terms of squares:

a = [121, 144, 19, 161, 19, 144, 19, 11]<br>
b = [11*11, 121*121, 144*144, 19*19, 161*161, 19*19, 144*144, 19*19]

Invalid arrays

If, for example, we change the first number to something else, comp is not returning true anymore:

a = [121, 144, 19, 161, 19, 144, 19, 11]<br>
b = [132, 14641, 20736, 361, 25921, 361, 20736, 361]

comp(a, b) returns false because in b, 132 is not the square of any number of a.

a = [121, 144, 19, 161, 19, 144, 19, 11]<br>
b = [121, 14641, 20736, 36100, 25921, 361, 20736, 361]

comp(a, b) returns false because in b, 36100 is not the square of any number of a.

Remarks

- a or b might be [] or {} (all languages except R, Shell).<br>
- a or b might be nil, null, None, or nothing depending on the language.<br>
- If a or b are nil (or null or None), the problem doesn't make sense, so return false.

```python
import math

def comp(array1, array2):
    if array1 is None or array2 is None:
        return False

    squared = []

    for num in array1:
        squared.append(num ** 2)

    return sorted(squared) == sorted(array2)
    return True
```

```python
comp([121, 144, 19, 161, 19, 144, 19, 11], [121, 14641, 20736, 361, 25921, 361, 20736, 361]) # True
```
