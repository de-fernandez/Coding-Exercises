Once upon a time, on a way through the old wild mountainous west,…<br>
… a man was given directions to go from one point to another. The directions were "NORTH", "SOUTH", "WEST", "EAST". Clearly "NORTH" and "SOUTH" are opposite, "WEST" and "EAST" too.<br>
<br>
Going to one direction and coming back the opposite direction right away is a needless effort. Since this is the wild west, with dreadful weather and not much water, it's important to save yourself some energy, otherwise you might die of thirst!<br>
<br>
How I crossed a mountainous desert the smart way.<br>
The directions given to the man are, for example, the following (depending on the language):<br>
<br>
["NORTH", "SOUTH", "SOUTH", "EAST", "WEST", "NORTH", "WEST"].<br>
<br>
You can immediately see that going "NORTH" and immediately "SOUTH" is not reasonable, better stay in the same place! So the task is to give to the man a simplified version of the plan. A better plan in this case is simply:<br>
<br>
["WEST"]<br>
<br>
Other examples:<br>
In ["NORTH", "SOUTH", "EAST", "WEST"], the direction "NORTH" + "SOUTH" is going north and coming back right away.<br>
<br>
The path becomes ["EAST", "WEST"], now "EAST" and "WEST" cancel each other out, therefore, the final result is [].<br>
<br>
In ["NORTH", "EAST", "WEST", "SOUTH", "WEST", "WEST"], "NORTH" and "SOUTH" are not directly opposite but they become directly opposite after the reduction of "EAST" and "WEST" so the whole path is reducible to ["WEST", "WEST"].<br>
<br>
Task<br>
Write a function `dirReduc` which will take an array of strings and returns an array of strings with the needless directions removed (W ↔ E or S ↔ N side by side).<br>
<br>
Notes<br>
Not all paths can be made simpler. The path ["NORTH", "WEST", "SOUTH", "EAST"] is not reducible. "NORTH" and "WEST", "WEST" and "SOUTH", "SOUTH" and "EAST" are not directly opposite of each other and can't become such. Hence the result path is itself: ["NORTH", "WEST", "SOUTH", "EAST"].

```python
def dirReduc(arr):
    result = []

    for direction in arr:
        if result:
            if (result[-1] == "NORTH" and direction == "SOUTH") or \
               (result[-1] == "SOUTH" and direction == "NORTH") or \
               (result[-1] == "EAST" and direction == "WEST") or \
               (result[-1] == "WEST" and direction == "EAST"):
                result.pop()
            else:
                result.append(direction)
        else:
            result.append(direction)

    return result
```

```python
dirReduc(["NORTH", "SOUTH", "SOUTH", "EAST", "WEST", "NORTH", "WEST"]) # WEST
```
