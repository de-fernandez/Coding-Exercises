Given a random non-negative number, you have to return the digits of this number within an array in reverse order.

Example (Input => Output):<br>
35231 => [1,3,2,5,3]<br>
0     => [0]

```python
def digitize(n):
    result = []
    for num in str(n):
        result.append(int(num))
    return result[::-1]
```

```python
digitize(23582357) # [7, 5, 3, 2, 8, 5, 3, 2]
```
