Complete the function that takes two integers (a, b, where a < b) and return an array of all integers between the input parameters, including them.

For example:

a = 1<br>
b = 4<br>
--> [1, 2, 3, 4]

```python
def between(a,b):
    arr = []
    for num in range (a, b + 1):
        arr.append(num)
    return arr
```

```python
between(1, 4) # [1, 2, 3, 4]
```
