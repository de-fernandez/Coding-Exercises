Given the triangle of consecutive odd numbers:

1 <br>
3     5<br>
7     9    11<br>
13    15    17    19<br>
21    23    25    27    29<br>
...<br>
Calculate the sum of the numbers in the nth row of this triangle (starting at index 1) e.g.: (Input --> Output)

1 -->  1<br>
2 --> 3 + 5 = 8

```python
def row_sum_odd_numbers(n):
    return n ** 3
```

```python
row_sum_odd_numbers(5)
```
