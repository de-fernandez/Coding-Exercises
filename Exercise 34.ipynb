Write a function that takes an array of numbers and returns the sum of the numbers. The numbers can be negative. If the array is empty, return 0.

Examples<br>
Input: [1, 5.2, 4, 0, -1]<br>
Output: 9.2

Input: [-2.398]<br>
Output: -2.398<br>

Input: []<br>
Output: 0

Assumptions<br>
You can assume that you are given a (possibly empty) valid array containing only numbers.

What We're Testing<br>
We're testing basic loops and math operations. This is for beginners who are just learning loops and math operations.<br>
Advanced users may find this extremely easy and can easily write this in one line.



```python
def sum_array(a):
    result = 0
    for x in a:
        result += x
    return result
```

```python
sum_array([1.1, 2.2, 3.3]) # 6.6
```
