In JavaScript, if..else is the most basic conditional statement, it consists of three parts: condition, statement1, statement2, like this:<br>
<br>
if condition: statementa<br>
else:         statementb<br>
<br>
It means that if the condition is true, then execute the statementa, otherwise execute the statementb. If the statementa or statementb are more than one line, you need to add { and } at the head and tail of statements in JS, to keep the same indentation on Python and to put an end in Ruby where it indeed ends.<br>
<br>
For example, if we want to judge whether a number is odd or even, we can write code like this:<br>
<br>
def odd_even(n):<br>
&nbsp;&nbsp;&nbsp;&nbsp;if n % 2: return "odd number"<br>
&nbsp;&nbsp;&nbsp;&nbsp;else:     return "even number"<br>
<br>
If there is more than one condition to judge, we can use the compound if...else statement. For example:<br>
<br>
def old_young(age):<br>
&nbsp;&nbsp;&nbsp;&nbsp;if age < 16:        return "children"<br>
&nbsp;&nbsp;&nbsp;&nbsp;elif age < 50:      return "young man"<br>
&nbsp;&nbsp;&nbsp;&nbsp;else:               return "old man"<br>
<br>
This function returns a different value depending on the parameter age.<br>
<br>
Looks very complicated? Well, JS and Ruby also support the ternary operator and Python has something similar too:<br>
<br>
statementa if condition else statementb<br>
<br>
The two examples above can be simplified with ternary operator:<br>
<br>
def odd_even(n):<br>
&nbsp;&nbsp;&nbsp;&nbsp;return "odd number" if n % 2 else "even number"<br>
<br>
def old_young(age):<br>
&nbsp;&nbsp;&nbsp;&nbsp;return "children" if age < 16 else "young man" if age < 50 else "old man"<br>
