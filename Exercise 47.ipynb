Clock shows h hours, m minutes and s seconds after midnight.

Your task is to write a function which returns the time since midnight in milliseconds.

Example:

h = 0<br>
m = 1<br>
s = 1

result = 61000

Input constraints:

0 <= h <= 23<br>
0 <= m <= 59<br>
0 <= s <= 59

```python
def past(h, m, s):
    return (h * 3600000) + (m * 60000) + (s * 1000)
```

```python
past(1,1,1) # 3661000
```
