The first input array is the key to the correct answers to an exam, like ["a", "a", "b", "d"]. The second one contains a student's submitted answers.

The two arrays are not empty and are the same length. Return the score for this array of answers, giving +4 for each correct answer, -1 for each incorrect answer, and +0 for each blank answer, represented as an empty string.

If the score < 0, return 0.

Examples:

Correct answers:  ["a", "a", "b", "b"]<br>
Student answers:  ["a", "c", "b", "d"]<br>
Output: 6

Correct answers:  ["a", "a", "c", "b"]<br>
Student answers:  ["a", "a", "b", ""]<br>
Output: 7

Correct answers:  ["a", "a", "b", "c"]<br>
Student answers:  ["a", "a", "b", "c"]<br>
Output: 16

Correct answers:  ["b", "c", "b", "a"]<br>
Student answers:  ["", "a", "a", "c"]<br>
Output: 0

```python
def check_exam(arr1, arr2):
    score = 0
    i = 0

    while i < len(arr1):
        if arr2[i] == "":
            score += 0
        elif arr1[i] == arr2[i]:
            score += 4
        else:
            score -= 1

        i += 1

    if score < 0:
        return 0

    return score
```

```python
check_exam(["a", "a", "b", "b"], ["a", "c", "b", "d"]) # 6
```
