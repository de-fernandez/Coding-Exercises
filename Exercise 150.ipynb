Take an integer n (n >= 0) and a digit d (0 <= d <= 9) as an integer.

Square all numbers k (0 <= k <= n) between 0 and n.

Count the numbers of digits d used in the writing of all the k**2.

Implement the function taking n and d as parameters and returning this count.

Examples:<br>
n = 10, d = 1<br>
the k*k are 0, 1, 4, 9, 16, 25, 36, 49, 64, 81, 100<br>
We are using the digit 1 in: 1, 16, 81, 100. The total count is then 4.<br>

The function, when given n = 25 and d = 1 as argument, should return 11 since<br>
the k*k that contain the digit 1 are:<br>
1, 16, 81, 100, 121, 144, 169, 196, 361, 441.<br>
So there are 11 digits 1 for the squares of numbers between 0 and 25.<br>
Note that 121 has twice the digit 1.<br>

```python
def nb_dig(n, d):
    total = 0

    for i in range(n + 1):
        square = i * i
        square = str(square)

        total += square.count(str(d))

    return total
```

```python
nb_dig(8304, 7) # 3927
```
