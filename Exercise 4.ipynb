Given a non-empty array of integers, return the result of multiplying the values together in order.

Example:

[1, 2, 3, 4] => 1 * 2 * 3 * 4 = 24

```python
def grow(arr):
    result = 1
    for x in range(0, len(arr)):
        result = result * arr[x]
    return result
```

```python
arr = [1, 2, 3]
grow(arr)
```

```python
arr = [1, 2, 3, 4, 5]
grow(arr)
```
