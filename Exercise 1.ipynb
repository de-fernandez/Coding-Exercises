Complete the solution so that it reverses the string passed into it.

'world'  =>  'dlrow'<br>
'word'   =>  'drow'

```python
def solution(string):
    placeholder = ""
    for x in range(0, len(string)):
        placeholder = string[x] + placeholder
    return placeholder
```

```python
solution("world")
```

```python
solution("word")
```
