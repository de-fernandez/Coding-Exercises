Rules of the "Rock, Paper, Scissors" game are:<br>

Rock beats Scissors<br>
Scissors beat Paper<br>
Paper beats Rock<br>
Two identical moves are a draw.<br>

Let's play! You will be given valid moves of two Rock, Paper, Scissors players, and have to return which player won: "Player 1 won!" for player 1, and "Player 2 won!" for player 2. In case of a draw return Draw!.<br>

Examples:<br>
"scissors",     "paper"     --> "Player 1 won!"<br>
"scissors",     "rock"      --> "Player 2 won!"<br>
"paper",        "paper"     --> "Draw!"

```python
def rps(p1, p2):
    choices = {"rock": 0, "paper": 1, "scissors": 2}
    p1 = choices[p1]
    p2 = choices[p2]

    if p1 == p2:
        return "Draw!"
    elif (p1 - p2) % 3 == 1:
        return "Player 1 won!"
    else:
        return "Player 2 won!"
```

```python
rps("rock", "scissors")
```
