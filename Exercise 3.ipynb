Remove First and Last Character<br>
Task<br>
Your goal is to write a function that removes the first and last characters of a string. You're given one parameter, the original string.

Important: Your function should handle strings of any length â‰¥ 2 characters. For strings with exactly 2 characters, return an empty string.

Examples
'eloquent' --> 'loquen'<br>
'country'  --> 'ountr'<br>
'person'   --> 'erso'<br>
'ab'       --> '' (empty string)<br>
'xyz'      --> 'y'<br>

Requirements<br>
The input string will always have at least 2 characters<br>
For strings with exactly 2 characters, return an empty string<br>
For strings with 3 or more characters, remove the first and last character<br>
The function should handle strings containing letters, numbers, and special characters<br>
Test Cases<br>

Your solution will be tested against:<br>
Basic functionality with common words<br>
Edge cases with 2-character and 3-character strings<br>
Strings containing numbers and special characters<br>
Random test cases of varying lengths<br>


```python
def remove_char(string):
    chars = list(string)
    chars.pop(0)
    chars.pop(-1)
    return "".join(chars)
```

```python
remove_char("Python")
```
