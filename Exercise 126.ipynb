Count the number of divisors of a positive integer n.

Random tests go up to n = 500000, but fixed tests go higher.

Examples (input --> output)<br>
4 --> 3 // we have 3 divisors - 1, 2 and 4<br>
5 --> 2 // we have 2 divisors - 1 and 5<br>
12 --> 6 // we have 6 divisors - 1, 2, 3, 4, 6 and 12<br>
30 --> 8 // we have 8 divisors - 1, 2, 3, 5, 6, 10, 15 and 30<br>
Note you should only return a number, the count of divisors. The numbers between parentheses are shown only for you to see which numbers are counted in each case.



```python
def divisors(n):
    count = 0

    for i in range(1, n + 1):
        if n % i == 0:
            count += 1

    return count
```

```python
divisors(55) # [1, 5, 11, 55]
```
