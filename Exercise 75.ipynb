Complete the solution so that the function will break up camel casing, using a space between words.

Example<br>
"camelCasing"  =>  "camel Casing"<br>
"identifier"   =>  "identifier"<br>
""             =>  ""

```python
def solution(s):
    result = ""
    for char in s:
        if char.isupper():
            result += " " + char
        else:
            result += char
    return result
```

```python
solution("helloWorld") # "hello World"
```
