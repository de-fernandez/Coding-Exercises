Task:<br>
Given a non-negative integer, 3 for example, return a string with a murmur: "1 sheep...2 sheep...3 sheep...". Input will always be valid, i.e. no negative integers.

```python
def count_sheep(n):
    result = ""
    for x in range(1, n + 1):
        result += f"{x} sheep..."
    return result
```

```python
count_sheep(3)
```

```python
count_sheep(8)
```
