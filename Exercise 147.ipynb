Write a function that returns a string in which firstname is swapped with last name.

Example(Input --> Output)

"john McClane" --> "McClane john"

```python
def name_shuffler(str_):
    words = str_.split()

    first = words[0]
    last = words[1]

    return last + " " + first
```

```python
name_shuffler('tom jerry') # jerry tom
```
